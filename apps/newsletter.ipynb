````python
# Request
def request_api(url):
    r = requests.get(url)
    json_data = r.json()
    keys = json_data.keys()

    if 'results' in keys:
        df = pd.DataFrame(json_data['results'])

    if 'response' in keys:
        df = pd.DataFrame(json_data['response']['docs'])
    
    return df
````

# Newsletter Generation with OpenAI

````python
import os
import openai
from dotenv import load_dotenv, find_dotenv
_ = load_dotenv(find_dotenv()) # read local .env file

openai.api_key  = os.getenv('OPENAI_API_KEY')
````

````python
def get_completion(prompt, model="gpt-3.5-turbo"): 
    messages = [{"role": "user", "content": prompt}]
    response = openai.ChatCompletion.create(
        model=model,
        messages=messages,
        temperature=0, # this is the degree of randomness of the model's output
    )
    return response.choices[0].message["content"]
````

````python
prompt = f"""
Your task is to generate a newsletter from the text below. \
The text is a list of articles (only tittle and abstract) from the New York Times bussiness section. \
Please analysis the sentiment of each article, and generate a newsletter in HTML format with the following structure: \
Newsletter of NYT \
<tittle> \
<sentiment> \
<abstract>\

News text: ```{text}```
"""

response = get_completion(prompt)
print(response)
````
